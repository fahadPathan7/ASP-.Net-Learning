#### with method

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printInfo(self):
        print(f"Name is {self.name}.\nAge is {self.age}")

    def salary(self, salary):
        print(f"{self.name}'s salary is: {salary}")

person1 = Person("Fahad", "23")
person1.printInfo()
print(person1.name)
person1.salary("35000")
```

### inheritance --->

```python
class Vehicle:
    def __init__(self):
        print("Hi, I am vehicle")
    def wheels(self):
        print("Vehicle has four wheels")

class Truck(Vehicle):
    def whistle(self):
        print("Truck wistle")

class Cycle(Vehicle):
    def __init__(self):
        print("I am cycle")
        
    def wheels(self):
        print("Cycle has 2 wheels")

truck = Truck()
truck.wheels()
cycle = Cycle()
cycle.wheels()
```

### encapsulation --->

```python
class Bank:
    def __init__(self, account, balance):
        self.__account = account
        self.__balance = int(balance)
    def addBalance(self, amount):
        self.__balance += int(amount)
    def getBalance(self):
        return self.__balance

bank = Bank("123", 1000)
bank.addBalance(500)
bank.getBalance()
```

### magic dunder method --->

```python
class Bank:
    def __init__(self, account, balance):
        self.__account = account
        self.__balance = int(balance)
    def __add__(self, amount):
        self.__balance += int(amount)
    def getBalance(self):
        return self.__balance

bank = Bank("123", 1000)
bank + 100
bank.getBalance()
```
